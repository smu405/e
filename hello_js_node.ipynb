* 상대경로 vs 절대경로
    * 로컬 파일은 보안정책으로 인해 읽을 수가 없다 Cross-Origin Restriction Policy

* 절대경로 
    * http://domain/ or just / (replace domain with your actual domain, such as www.example.com).
* 상대경로
    * 현재 HTML page로부터의 경로

#### main 찾기

그 라이브러리가 설치된 디렉토리에서 ```package.json```을 읽어서 ```main```을 찾는다. 
예를 들어, ```./node_modules/web3```를 찾으면 그 아래 다음과 같이 ```package.json```이 저장되어 있고, 그 파일을 열어보면 ```main```이 무엇인지 확인할 수 있다..

```python
node_modules/web3
    |- package.json  이 파일을 열어보면, main이 src/index.js라고 적혀있다.
    |- src/
        |- index.js
```

```python
{
  "name": "web3",
  ...
  "version": "1.2.5-rc.0",
  ...
  "main": "src/index.js"  package.json의 main 필드에 src/index.js가 적혀 있다.
}
```

따라서 ```require('web3.js')```라고 하면, ```main```의 ```src/index.js```가 반환된다.
```index.js```에 ```module.exports=Web3``` 등 필요한 필요한 설정이 적혀있다.
```package.json```에 ```main```이 명시되어 있지 않으면, ```index.js```가 그냥 쓰이게 된다.

#### 간단한 exports

**상대경로**는 ./ 또는 ../를 사용하는 경우를 말하는데, 현재 파일을 기준으로 상대적으로 require()의 경로로 넣으면 된다.
예를 들어, ```src/testImport.js```에서 ```src/test.js```를 호출한다면, ```require('./src/test.js')```가 아닌 ```require('./test.js')```라고 써준다.

````python
%%writefile src/test.js
exports.DateOfBirth = '11.11.1999';
````

````python
%%writefile src/testImport.js
const my = require('./test');
console.log(my)
````
